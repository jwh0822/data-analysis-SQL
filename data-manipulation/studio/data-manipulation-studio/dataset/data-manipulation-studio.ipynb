## Getting Started

Import the CSVs for each of the following cities: Baltimore, Boston, New York, and Philadelphia. Set up a dataframe for each city.

```python
# Import the necessary libraries and CSVs. Make some dataframes!
import pandas as pd
import numpy as np

balt_data = pd.read_csv("baltimore_9-24-2016_9-30-2017.csv")
boston_data = pd.read_csv("boston_9-24-2016_9-30-2017.csv")
ny_data = pd.read_csv("new-york_9-24-2016_9-30-2017.csv")
phil_data = pd.read_csv("philadelphia_9-24-2016_9-30-2017.csv")

chicago_data = pd.read_csv("chicago_9-24-2016_9-30-2017.csv")
detroit_data = pd.read_csv("detroit_9-24-2016_9-30-2017.csv")
stl_data = pd.read_csv("st-louis_9-24-2016_9-30-2017.csv")
```

## Clean Your Data

In the last lesson, we cleaned the data related to San Francisco. Pull up your notebook from the last lesson and use it as a reference to clean up these new dataframes.

```python
# Clean your data here!
col_list = ['Origin District', 'Type', 'Environment', 'Quality', 'Condition', 'Grade', 'Appearance', 'Storage', 'Crop', 'Color', 'Trans Mode']

balt_data = balt_data.drop(labels = col_list, axis=1)
boston_data = boston_data.drop(labels = col_list, axis=1)
ny_data = ny_data.drop(labels = col_list, axis=1)
phil_data = phil_data.drop(labels = col_list, axis=1)

chicago_data = chicago_data.drop(labels = col_list, axis=1)
detroit_data = detroit_data.drop(labels = col_list, axis=1)
stl_data = stl_data.drop(labels = col_list, axis=1)
```

## Combine Your Data

Now that you have four clean sets of data, combine all four into one dataframe that represents the entire Northeast region.

```python
# Combine the four dataframes into one!
northeastregion = pd.concat([balt_data, boston_data, ny_data, phil_data])
northeastregion.tail()
```

## Answer Some Questions

Use `groupby()` and `agg()` to answer the following two questions:

1. What is the mean low and high prices for each type of **unit of sale** in the Northeast region? 
2. For each region, what is the average number of pumpkins per variety that came into terminal markets for the year? 

```python
# Put your code here to find the mean low and high prices in the Northeast region for each type of unit of sale.
northeastregion.groupby("Unit of Sale").agg(mean_low=("Low Price", 'mean'), mean_high=("High Price", 'mean'))
```

```python
# Put your code here to find the average number of pumpkins coming into terminal markets of each variety.
northeastregion.groupby("City Name").count().groupby("City Name").agg(avg_pumpkins_each_terminal=("Variety", 'mean'))
#northeastregion["Variety"].value_counts()
```
