2. Write a query that returns the book ids of all books that have over 1000 ratings of 1 star or over 1000 ratings of 5 stars.

```sql
/*USE BooksDB;
SELECT dbo.books.book_id
FROM dbo.books
WHERE dbo.books.book_id IN (
    SELECT r.book_id
    FROM dbo.ratings AS r
    WHERE r.rating = 1 or r.rating = 5
    GROUP BY r.book_id
);*/
```

3. Write a query that returns the book ids of all books that have over 1000 ratings of 1 star and over 1000 ratings of 5 stars.

```sql
/*USE BooksDB;
SELECT DISTINCT book_id
FROM dbo.ratings
WHERE book_id IN (
    SELECT book_id
    FROM dbo.ratings
    GROUP BY book_id
    HAVING SUM(CASE WHEN rating = 1 THEN 1 END ) > 1000
    AND SUM(CASE WHEN rating = 5 THEN 1 END) > 1000
);*/
```
