### 4. Data Selection -- Indexing and Slicing

```python

#  Extract data as series
# define and print the variable 'genre'
genre_series = df.Genre
print(genre_series)
```

```python

# Extract the data as a dataframe and print it
my_df = pd.DataFrame(df)
print(my_df)

```

```python
# Extract data using rows

# loc function
print(df.loc[99])
# the loc method is used to locate rows or columns based on label-indexing
```

```python
# iloc function
print(df.iloc[255])
# the iloc method is used to locate rows or columns based on index position.
```

### 5. Data Selection - Based on Conditional Filtering

```python
# Use the code block to select movies that have been released between 2010-2016, with a rating of less than 6.0, but generated a signifianct (top) revenue.
movies = my_df[(my_df['Year'] >= 2010) & (my_df["Year"] <= 2016)]
ratings = my_df[(my_df['Rating'] < 6.0)]
movies_ratings = movies + ratings
print(movies_ratings)
```
